# Фазовая диаграмма физического маятника без начальной скорости

```python
import numpy as np
import matplotlib.pyplot as plt

with open('Ae.bin', 'rb') as file:
    Ae = np.fromfile(file, dtype=float, count=-1)

with open('Be.bin', 'rb') as file:
    Be = np.fromfile(file, dtype=float, count=-1)
    
plt.plot(Ae, Be, label=r'$Euler$')

with open('Ah.bin', 'rb') as file:
    Ah = np.fromfile(file, dtype=float, count=-1)

with open('Bh.bin', 'rb') as file:
    Bh = np.fromfile(file, dtype=float, count=-1)
    
plt.plot(Ah, Bh, label=r'$Heun$')
plt.legend(loc='best', fontsize=12)
```
