# Фигуры Лиссажу

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_Lissajous (A, B, delta, omega_1, omega_2):
    lcm = int(2 * np.pi / omega_1) * int(2 * np.pi / omega_2)
    t = np.linspace(0.0, lcm, 1000)
    x = A * np.sin(omega_1 * t + delta)
    y = B * np.cos(omega_2 * t)
    return plt.plot(x, y)

delta = [np.pi / 4, np.pi / 6, np.pi / 8, np.pi / 3]
omega_1 = [2 * np.pi / 3, np.pi, np.pi / 4, 2 * np.pi / 5]
omega_2 = [2 * np.pi / 5, 2 * np.pi / 3, 2 * np.pi / 13, 2 * np.pi / 9]
A = [50, 40, 30, 70]
B = [40, 70, 60, 30]


fig, axs = plt.subplots(2, 2)
for i in range(4):
    sp = plt.subplot(2, 2, i + 1)
    draw_Lissajous(A[i], B[i], delta[i], omega_1[i], omega_2[i])
plt.show()
```

# Гистограмма распределения абсцисс и ординат векторов фигуры Лиссажу

```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt

omega_1 = 2 * np.pi / 3
omega_2 = 2 * np.pi / 5
A = 50
B = 40
delta = np.pi / 4
lcm = int(2 * np.pi / omega_1) * int(2 * np.pi / omega_2)
t = np.linspace(0.0, lcm, 1000)
x = A * np.sin(omega_1 * t + delta)
y = B * np.cos(omega_2 * t)

sp = plt.subplot(2, 1, 1)
plt.title('Histogram of x')
plt.xlabel('values')
plt.ylabel('frequency')
plt.hist(x, 50, density=True)

sp = plt.subplot(2, 1, 2)
plt.title('Histogram of y')
plt.xlabel('values')
plt.ylabel('frequency')
plt.hist(y, 50, density=True)

plt.show()
```

# МНК

```python
import numpy as np
import matplotlib.pyplot as plt

with open ('x.txt') as fx:
  x0 = [[float(number) for number in line.split()] for line in fx]


with open ('y.txt') as fy:
  y0 = [[float(number) for number in line.split()] for line in fy]

x = numpy.empty(np.size(x0), float)
y = numpy.empty(np.size(y0), float)
for i in range (np.size(x0)):
    x[i] = x0[i][0]
    
for i in range (np.size(y0)):
    y[i] = y0[i][0]


    
plt.plot(x, y, 'ro')
p, v = np.polyfit(x, y, deg=1)
x_fit = np.linspace(0.00, 0.23, 100)
plt.plot(x_fit, p * x_fit + v)
```
