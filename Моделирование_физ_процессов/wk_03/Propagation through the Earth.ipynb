```python
import numpy  as np
import matplotlib.pyplot as plt

R = 6371 * 10**6
a = np.empty(10000, float)
b = np.empty(10000, float)
t = np.linspace(0.0, 100, 10000)

a[0] = R
b[0] = 0.0
dt = 0.01

for i in range (9999):
    a[i+1] = a[i] + b[i] * dt
    b[i+1] = b[i] - a[i] * dt
```

# Фазовая диаграмма 

```python
plt.plot(a, b)
```

# Координата от времени 

```python
plt.plot(t, a)
```

# Скорость от времени

```python
plt.plot(t, b)
```

# Энергия от времени

```python
density = 5520
energy = np.empty(10000, float)

for i in range (10000):
    energy[i] = (b[i]**2) / 2 + (a[i]**2) / 2
    
plt.plot(t, energy)
```

# Фазовая диаграмма (C++) для разных dt

```python
import numpy  as np
import matplotlib.pyplot as plt

with open ('a1.txt') as fx1:
  a01 = [[float(number) for number in line.split()] for line in fx1]

with open ('b1.txt') as fy1:
  b01 = [[float(number) for number in line.split()] for line in fy1]

with open ('a2.txt') as fx2:
  a02 = [[float(number) for number in line.split()] for line in fx2]

with open ('b2.txt') as fy2:
  b02 = [[float(number) for number in line.split()] for line in fy2]

a1 = np.empty(10, float)
b1 = np.empty(10, float)

a2 = np.empty(100, float)
b2 = np.empty(100, float)

for i in range (10):
    a1[i] = a01[i][0]
    
for i in range (10):
    b1[i] = b01[i][0]
    
for i in range (100):
    a2[i] = a02[i][0]
    
for i in range (100):
    b2[i] = b02[i][0]
    
plt.plot(a1, b1)
plt.plot(a2, b2)
```

# Координата от времени (С++) для разных dt

```python
t1 = np.linspace(0.0, 1, 10)
t2 = np.linspace(0.0, 1, 100)

plt.plot(t1, a1)
plt.plot(t2, a2)
```

# Скорость от времени (С++) для разных dt

```python
plt.plot(t1, b1)
plt.plot(t2, b2)
```

# Энергия от времени (C++) для разных dt 

```python
plt.plot(t1, (a1**2)/2 + (b1**2)/2)
plt.plot(t2, (a2**2)/2 + (b2**2)/2)
```
