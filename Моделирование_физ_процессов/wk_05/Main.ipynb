# Энергия в момент t_0 от dt (float)

```python
import numpy  as np
import matplotlib.pyplot as plt

with open ('e float.txt') as fe:
  e0 = [[float(number) for number in line.split()] for line in fe]

e = np.empty(8, float)

for i in range (8):
    e[i] = e0[i][0]
    
dt = [-1, -2, -3, -4, -5, -6, -7, -8]

plt.xlabel(r'$\lg dt$')
plt.ylabel(r'$Eenergy(t = 1 s)$')

plt.scatter(dt, e)
```

# Фазовая диаграмма Эйлер vs Хойн

```python
import numpy  as np
import matplotlib.pyplot as plt

with open ('ae.txt') as fe:
  ae0 = [[float(number) for number in line.split()] for line in fe]

ae = np.empty(10000, float)

for i in range (10000):
    ae[i] = ae0[i][0]

with open ('be.txt') as fe:
  be0 = [[float(number) for number in line.split()] for line in fe]

be = np.empty(10000, float)

for i in range (10000):
    be[i] = be0[i][0]

with open ('ac.txt') as fe:
  ac0 = [[float(number) for number in line.split()] for line in fe]

ac = np.empty(10000, float)

for i in range (10000):
    ac[i] = ac0[i][0]

with open ('bc.txt') as fe:
  bc0 = [[float(number) for number in line.split()] for line in fe]

bc = np.empty(10000, float)

for i in range (10000):
    bc[i] = bc0[i][0]

plt.plot(ae, be, label=r'$Euler$')
plt.plot(ac, bc, label=r'$Heun$')
plt.legend(loc='best', fontsize=12)
```
